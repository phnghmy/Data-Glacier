# Section 2: Problem Description

ABC Bank aims to develop a machine learning model to predict whether a customer will subscribe to a term deposit product. This model will help the bank focus its marketing efforts on customers with a higher likelihood of purchasing the product, thereby optimizing resource allocation and reducing marketing costs. The bank has provided two datasets: "bank-full" and "bank-additional-full," containing various customer attributes and campaign information.

# Section 5: Recommended Models (Technical Users)

Slide 11 (Technical Users): Recommended Models

Based on the EDA and feature engineering, we recommend the following models:

1. Logistic Regression: Baseline model for interpretability and key feature identification.
2. Random Forest Classifier: To capture complex feature interactions and non-linear relationships.
3. XGBoost or LightGBM: High performance on tabular data, handles imbalanced datasets well.
4. Neural Networks: For capturing intricate patterns and deep feature interactions.

Model Development Strategy:
- Address class imbalance using SMOTE
- Perform feature selection based on importance scores
- Use cross-validation for robust performance estimation
- Optimize hyperparameters using grid search or Bayesian optimization
- Evaluate models using AUC-ROC, precision, recall, and F1-score
- Interpret model results using SHAP values for actionable insights
