```python
X_encoded = pd.get_dummies(X, columns=['Gender', 'Payment_Mode'])
X_encoded
```

```python
Y_encoded = pd.get_dummies(Y, columns=['Company'])
Y_encoded
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, Y_encoded, test_size=0.2, random_state=42)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, max_depth = 3, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
predictions = rf_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, predictions)
precision = precision_score(y_test, predictions, average='weighted')
recall = recall_score(y_test, predictions, average='weighted')
f1 = f1_score(y_test, predictions, average='weighted')

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```

```python
# Extract a single tree from the forest
estimator = rf_model.estimators_[0]

# Export the tree as dot file
plot = export_graphviz(estimator, out_file='tree.dot', 
                feature_names=X_train.columns.tolist(),
                class_names=y_train.columns.tolist(),
                rounded=True, proportion=False, 
                precision=2, filled=True)
```

```python
try:
    dot_data = export_graphviz(estimator, out_file=None, 
                               feature_names=X_train.columns,
                               class_names=y_train.columns,
                               rounded=True, proportion=False, 
                               precision=2, filled=True)
    # Check if dot_data is None
    if dot_data is None:
        print("dot_data is None. There might be an issue with the export_graphviz parameters.")
    else:
        # Create graph from dot data
        graph = graphviz.Source(dot_data)
        
        # Display the tree plot inline
        display(graph)
except Exception as e:
    print(f"An error occurred: {e}")
```
